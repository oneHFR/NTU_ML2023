# Modify the generated image name
reference link
https://blog.csdn.net/weixin_44354586/article/details/103784884

https://blog.csdn.net/weixin_42370340/article/details/102833798

```python

import os
path = '/kaggle/working/results/predict/default'
filelist = os.listdir(path)
filelist.sort()
total_num = len(filelist)
print(total_num)
i = 0
```

```python
#reference https://blog.csdn.net/weixin_44354586/article/details/103784884
i=1
for item in filelist:
    if item.endswith('.jpg'):
        src = os.path.join(os.path.abspath(path), item)
        s = str(i)
        if i>1000:
            #reference https://blog.csdn.net/weixin_42370340/article/details/102833798
            os.remove(src)
        else:
            dst = os.path.join(os.path.abspath(path), s + '.jpg')
            os.rename(src, dst)
        i = i + 1
```
